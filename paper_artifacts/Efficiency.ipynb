```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 17})
```

```python
map_ne = [0.825, 0.7986189479, 0.844668, 0.773622488]
map_gmn = [0.622, 0.6413473489, 0.6785019635, 0.6512850003]

time_ne = [61.9, 78.7, 95.2, 76.5]
time_gmn = [42, 66.3, 82.5, 98.7]

labels_ne = ["(T=3, K=5)", "(T=4, K=5)", "(T=5, K=5)", "(T=3, K=10)"]
labels_gmn = ["K=5", "K=8", "K=10", "K=12"]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4,3))
ax.scatter(time_ne, map_ne, color='r', s=50, label="EinsMatch (Node)")
ax.scatter(time_gmn, map_gmn, color='b', s=50, label="GMN-Match")

legend = plt.legend(loc='upper left', bbox_to_anchor=(-0.02, 1.05), prop={'size': 15}, handlelength=0.01, handletextpad=0.3)
legend.get_frame().set_alpha(0)


ax.set_ylim([0.6, 1])
ax.set_xlim([30, 110])

ax.set_yticks([0.6, 0.8, 1.0])
ax.set_xticks([40, 60, 80, 100])

for i, txt in enumerate(labels_ne):
  if i == 3:
    plt.text(time_ne[i] + 0.02, map_ne[i] - 0.04, r"$\bf{" + txt + "}$", ha="right", fontsize=10)
  elif i == 1:
    plt.text(time_ne[i] + 0.01, map_ne[i] + 0.02, r"$\bf{" + txt + "}$", ha="center", fontsize=10)
  elif i == 0:
    plt.text(time_ne[i] + 0.1, map_ne[i] + 0.02, r"$\bf{" + txt + "}$", ha="right", fontsize=10)
  else:
    plt.text(time_ne[i] - 0.01, map_ne[i] + 0.02, r"$\bf{" + txt + "}$", ha="center", fontsize=10)

for i, txt in enumerate(labels_gmn):
  plt.text(time_gmn[i] - 0.01, map_gmn[i] + 0.01, r"$\bf{" + txt + "}$", ha="center", fontsize=10)

ax.set_xlabel('Inference time for 128 graph pairs\n(in milliseconds)', fontsize=13)
ax.set_ylabel('MAP', fontsize=20)

plt.savefig("efficiency.pdf", bbox_inches='tight')  # Add bbox_inches='tight' to adjust the boundaries automatically
```
