```python
import pickle
import seaborn as sns
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
# latexify
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('axes', linewidth=1)
plt.rc('font', weight='bold')

matplotlib.rcParams.update(matplotlib.rcParamsDefault)
matplotlib.rc('text', usetex=True)
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['axes.spines.right'] = False
matplotlib.rcParams['axes.spines.top'] = False
matplotlib.rcParams['text.latex.preamble']=r"\usepackage{bm} \usepackage{amsmath,amsfonts} \boldmath"
```

```python
histogram_dump_0 = {}
histogram_dump_1 = {}
histogram_dump_2 = {}
histogram_dump_3 = {}
histogram_dump_4 = {}

for dataset in ["aids", "mutag", "ptc_fm", "ptc_fr", "ptc_mm", "ptc_mr"]:
    with open(f"model_time_{dataset}_0", "rb") as f:
        histogram_dump_0[dataset] = pickle.load(f)

    with open(f"model_time_{dataset}_1", "rb") as f:
        histogram_dump_1[dataset] = pickle.load(f)

    with open(f"model_time_{dataset}_2", "rb") as f:
        histogram_dump_2[dataset] = pickle.load(f)

    with open(f"model_time_{dataset}_3", "rb") as f:
        histogram_dump_3[dataset] = pickle.load(f)

    with open(f"model_time_{dataset}_4", "rb") as f:
        histogram_dump_4[dataset] = pickle.load(f)
```

```python
plt.rcParams.update({'font.size': 17})
```

```python
latex_dataset={
    'aids':r'\textbf{AIDS}', 'mutag':r'\textbf{MUTAG}',
    'ptc_fm':r'\textbf{PTC-FM}', 'ptc_fr':r'\textbf{PTC-FR}',
    'ptc_mm':r'\textbf{PTC-MM}', 'ptc_mr':r'\textbf{PTC-MR}'
}

i=0
for dataset in ["aids", "mutag", "ptc_fm", "ptc_fr", "ptc_mm", "ptc_mr"]:
  if i in [0, 1, 2]:
    fig, (data_plt) = plt.subplots(1, 1, figsize=(4,3))
  else:
    fig, (data_plt) = plt.subplots(1, 1, figsize=(3.8,2.8))

  plt.subplots_adjust(wspace=0.1)

  if i in [0]:
    top_needed = 0.2
  elif i in [2, 5]:
    top_needed = 0.22
  elif i in [3, 1]:
    top_needed = 0.18
  else:
    top_needed = 0.16

  binwidth=0.1
  alpha_=0.4
  lw_=0.8

  total = np.concatenate((histogram_dump_0[dataset], histogram_dump_1[dataset], histogram_dump_2[dataset], histogram_dump_3[dataset], histogram_dump_4[dataset]))

  sns.histplot(histogram_dump_0[dataset], kde=False, stat="probability", color='r', element="step", label=latex_dataset[dataset],
          binwidth=binwidth, ax=data_plt, alpha=0.4, binrange=(0, np.ceil(np.max(total))), palette='pastel', lw=lw_)
  sns.histplot(histogram_dump_1[dataset], kde=False, stat="probability", color='b', element="step",
          binwidth=binwidth, ax=data_plt, alpha=0.3, binrange=(0, np.ceil(np.max(total))), palette='pastel', lw=lw_)
  sns.histplot(histogram_dump_2[dataset], kde=False, stat="probability", color='g', element="step",
          binwidth=binwidth, ax=data_plt, alpha=0.3, binrange=(0, np.ceil(np.max(total))), palette='pastel', lw=lw_)
  sns.histplot(histogram_dump_3[dataset], kde=False, stat="probability", color='orange', element="step",
          binwidth=binwidth, ax=data_plt, alpha=0.3, binrange=(0, np.ceil(np.max(total))), palette='pastel', lw=lw_)
  sns.histplot(histogram_dump_4[dataset], kde=False, stat="probability", color='purple', element="step",
          binwidth=binwidth, ax=data_plt, alpha=0.3, binrange=(0, np.ceil(np.max(total))), palette='pastel', lw=lw_)

  data_plt.set(ylabel=None)
  top = data_plt.get_ylim()[1]

  data_plt.set_xticks([0, 5, 10])
  data_plt.set_xlim([0,np.ceil(np.max(total))])
  data_plt.set_xlabel(r'\boldmath$\text{Tr}(P^{\top} P^*)\rightarrow$', fontsize=25)

  data_plt.set_yticks([0.04, 0.08, 0.12, 0.16, 0.20, 0.24])

  if i in [0, 1, 2]:
    data_plt.set_ylabel(r'\boldmath$p(\text{Tr}(P^{\top} P^*))\rightarrow$', fontsize=25,labelpad=8)
  data_plt.set_ylim([0,top_needed])

  ax2 = data_plt.twinx()
  ax2.axes.yaxis.set_ticks([])

  alpha_=0.8
  lw_=2.5
  sns.kdeplot(histogram_dump_0[dataset], color="r", label="$k=1$", lw=lw_, bw_adjust=1, ax=ax2, alpha=alpha_)
  sns.kdeplot(histogram_dump_1[dataset], color="b", label="$k=2$", lw=lw_, bw_adjust=2, ax=ax2, alpha=alpha_)
  sns.kdeplot(histogram_dump_2[dataset], color="g", label="$k=3$", lw=lw_, bw_adjust=2, ax=ax2, alpha=alpha_)
  sns.kdeplot(histogram_dump_3[dataset], color="orange", label="$k=4$", lw=lw_, bw_adjust=2, ax=ax2, alpha=alpha_)
  sns.kdeplot(histogram_dump_4[dataset], color="purple", label="$k=5$", lw=lw_, bw_adjust=2, ax=ax2, alpha=alpha_)

  top_now = ax2.get_ylim()[1]
  ax2.set_ylim([0, top_now * top_needed / top])

  ax2.legend(bbox_to_anchor=(-0.04, -0.04, 1.08, 1.09), prop={'size': 20}, frameon=False, handlelength=0.4, labelspacing=0.1, handletextpad=0.2)
  if i in [0]:
    leg = data_plt.legend(bbox_to_anchor=(-0.38, -0.12, 1, 1.09),  prop={'size': 20}, frameon=False, handlelength=0, handletextpad=-0.2)
  elif i in [1]:
    leg = data_plt.legend(bbox_to_anchor=(-0.33, -0.12, 1, 1.09),  prop={'size': 20}, frameon=False, handlelength=0, handletextpad=-0.2)
  elif i in [1, 2, 3]:
    leg = data_plt.legend(bbox_to_anchor=(-0.3, -0.12, 1, 1.09),  prop={'size': 20}, frameon=False, handlelength=0, handletextpad=-0.2)
  else:
    leg = data_plt.legend(bbox_to_anchor=(-0.29, -0.12, 1, 1.09),  prop={'size': 20}, frameon=False, handlelength=0, handletextpad=-0.2)

  for item in leg.legendHandles:
      item.set_visible(False)

  plt.box(on=True)
  data_plt.spines['top'].set_visible(True)
  data_plt.spines['right'].set_visible(True)

  file = f"{dataset}_node_baseline_histogram.pdf"
  plt.savefig(file, bbox_inches='tight')
  plt.show()
  i += 1
```
